## Scatter Plots

##### Now we will use matplotlib to plot the graph between 3P Made and FT%

```python
plt.plot(data['3P Made'], data['FT%'], 'ro') # 'ro' represents scatter plot in matplotlib
```

##### Here we will use pandas directly to plot the graph between 3P Made and FT%

1. The advantage of pandas to plot is we can name both x and y axis at a time as column names and plot also looks better

```python
data.plot(kind = "Scatter", x = '3P Made', y = 'FT%') 
```


The user can define his own axis by using the command: a = plt.figure(figsize=(x_co_ordinates,y_co_ordinates))
                                                            
                                                       fit = a.gca()
                                                            
                                                       data.plot(kind = "Scatter", x = "X_Column", y = "Y_Columns", ax=fit)

```python
ax = plt.figure(figsize=(6,6))
ax = ax.gca()
data.plot(kind = "Scatter", x = "3P Made", y = "FT%", ax=ax) # Here ax = ax represents to draw scatter in the block we created. 
plt.title("Sample Graph Scatter Plot")
```

# Line Plots

Line plot is a default plot in both matplotlib and as well as in pandas

```python
fig  = plt.figure(figsize=(8,8))
ax = fig.gca()
plt.plot(data['3P Made'],data['FT%'])
plt.title("Sample plot for line plot")
```

Plotting using pandas

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca()
data.plot(kind = "Line", x = "3P Made", y = "FT%", ax = ax)
plt.title("Sample plot of line plot using pandas")
```

# Using seaborn

1. Import Seaborn Package
2. Set the stlye of grid
3. Define Plot type with kdeplot

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid") # one of most used styles in data science
```

```python
sns.kdeplot(data['3P Made']) # This gives probability density curve
```

```python
sns.kdeplot(data['FT%']) # This also gives probability density curve
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca()
sns.set_style("whitegrid")
sns.kdeplot(data["3P Made"], ax=ax)
plt.title("Sample Seaborn")
```

```python
sns.set_style("darkgrid")
sns.sinplot(data["3P Made"], np.sin())
```
